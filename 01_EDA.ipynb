# 01. Exploratory Data Analysis (EDA) — MIMIC-IV-ED

This notebook contains the initial data exploration of the MIMIC-IV Emergency Department dataset.  
The goal is to understand the structure, volume, and quality of each table before any preprocessing or modeling.

**Tables available:**
- `edstays` — Core table: one row per ED visit
- `triage` — Vital signs and chief complaint recorded at patient arrival
- `diagnosis` — ICD diagnosis codes assigned during the stay
- `medrecon` — Medication reconciliation (prior medications)
- `pyxis` — Medications dispensed inside the ED
- `vitalsign` — Repeated vital sign measurements throughout the stay

## 0. Setup

```python
import pandas as pd
import os

# Path to the data folder
data_path = r"mimic-iv-ed-2.2\ed"

print("Setup complete. Data path:", data_path)
```

## 1. Table-by-Table Exploration

For each table we will inspect:
- The first rows (`.head()`)
- Total number of rows
- Number of unique patients (`subject_id`)
- Number of unique stays (`stay_id`)
- Top 15 most frequent values of the most relevant categorical column

### 1.1 Diagnosis

```python
df_diagnosis = pd.read_csv(os.path.join(data_path, "diagnosis.csv.gz"))

print(f"Diagnosis table has {df_diagnosis.shape[0]:,} rows and {df_diagnosis.shape[1]} columns.\n")
display(df_diagnosis.head())

# Count of unique IDs
patients_diagnosis = df_diagnosis['subject_id'].nunique()
visits_diagnosis = df_diagnosis['stay_id'].nunique()
print(f"Unique patients: {patients_diagnosis:,}")
print(f"Unique emergency visits (Stays): {visits_diagnosis:,}\n")

# Top 15 most frequent diagnoses
print("Top 15 most common diagnoses (ICD Title):")
display(df_diagnosis['icd_title'].value_counts().head(15))
```

From the extracted data, the table contains a large number of entries (899,050 rows) belonging to a smaller number of unique patients (205,129) and unique hospital stays (423,989). This indicates that a single patient typically receives **multiple diagnoses during the same visit**.

## 3. Conclusions
### GEMINI
Basándonos en nuestra tabla comparativa de volumen, podemos inferir varias conclusiones fundamentales sobre la estructura de nuestra base de datos que dictarán cómo trataremos las variables antes de modelar:

1. **`Edstays` y `Triage` son la base perfecta (Core Tables):**
   Ambas tablas comparten exactamente la misma cantidad de filas (425,087), visitas únicas (425,087) y pacientes únicos (205,504). Esto nos garantiza matemáticamente que hay una relación de **1 a 1**; es decir, todo paciente que es admitido en urgencias pasa por un único triaje inicial obligatorio. *(`Edstays` será nuestra tabla "ancla" para calcular la variable objetivo de readmisión).*

2. **La Tasa Base de Pacientes Recurrentes:**
   Los `205,504` pacientes únicos generan un total de `425,087` estancias. Esto implica un promedio general de unas **2 visitas a urgencias por paciente**, lo cual nos adelanta que el fenómeno de la **readmisión** está fuertemente presente en nuestra base de datos.

3. **Múltiples diagnósticos por estancia (`Diagnosis`):**
   Aunque hay casi 424,000 estancias diagnosticadas, los registros se elevan a casi `900,000` filas. Esto confirma que el médico promedio suele asignar **más de un código ICD (diagnóstico)** a un paciente durante una misma visita a emergencias (por ejemplo, el motivo principal más una patología crónica).

4. **El historial farmacológico masivo (`Medrecon`):**
   Esta tabla posee un abrumador total de casi `3 millones` de filas para tan solo `307,196` visitas. Esto indica que los pacientes suelen llegar con listas de prescripciones previas bastante extensas (una media aproximada de 10 medicamentos crónicos por persona). Sin embargo, notablemente, **solo cubre a ~141,000 pacientes**, revelando que miles de personas acuden a urgencias sin historial previo de medicación.

5. **Dispensación clínica y Monitorización (`Pyxis` y `Vitalsign`):**
   Ambas tablas operan bajo un esquema de "múltiples tomas". Un paciente (un `stay_id`) recibe varias rondas de constantes vitales (`Vitalsign` reporta 1.5M de filas) y múltiples dispensaciones de medicamentos (`Pyxis`). Tendremos que agrupar (agregar usando promedios o *min/max*) estas filas por estancia antes de unirlas a nuestra tabla principal, porque de lo contrario duplicaríamos un mismo paciente cientos de veces en nuestro dataset de Machine Learning.

## CLAUDE

Based on the comparative table, we can draw the following key conclusions:

1. **`Edstays` and `Triage` are the core tables (1-to-1 relationship):**  
   Both share exactly the same number of rows (425,087), unique patients (205,504) and unique stays (425,087). Every admitted patient undergoes exactly one triage. `Edstays` will serve as our **anchor table** when building the readmission target variable.

2. **High patient recurrence rate:**  
   205,504 unique patients generate 425,087 total stays, meaning an average of ~2 ED visits per patient. This confirms that **readmission is a frequent and relevant phenomenon** in this dataset.

3. **Multiple diagnoses per stay (`Diagnosis`):**  
   899,050 rows for ~424,000 unique stays means physicians assign on average **more than 2 ICD codes per visit** (primary condition + chronic comorbidities).

4. **Large medication history (`Medrecon`):**  
   Nearly 3 million rows for ~307,000 stays indicates patients arrive with extensive prior prescription lists (~10 chronic medications on average). Notably, this table only covers ~141,000 patients, meaning many patients have **no prior medication history** recorded.

5. **Repeated measurements require aggregation (`Pyxis` and `Vitalsign`):**  
   Both tables contain multiple rows per stay (many dispensations and vital sign readings per visit). Before merging them into a master dataset for modeling, we will need to **aggregate these rows per `stay_id`** (e.g., using mean, min, max) to avoid duplicating patients.
